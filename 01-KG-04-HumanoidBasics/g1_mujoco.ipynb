## MuJoCo XML: what you’re looking at

MuJoCo models are defined in an **MJCF XML** file. At a high level, the XML describes:

- **Assets** (meshes, textures, materials) used for rendering: `<asset>`.
- **World & bodies** (the kinematic tree): `<worldbody>` with nested `<body>` elements.
- **Joints** that add degrees of freedom: `<joint>` (e.g., hinge, slide, ball).
- **Geometry / collision shapes** attached to bodies: `<geom>` (e.g., box, capsule, mesh).
- **Sites** (named frames/markers) useful for sensors, contacts, IK targets: `<site>`.
- **Actuators** that drive joints or bodies: `<actuator>` (e.g., motor, position, velocity).
- **Sensors** that expose signals (IMU, force/torque, joint pos/vel, etc.): `<sensor>`.
- **Simulation options** (timestep, solver parameters, gravity): `<option>`, `<size>`, `<default>`, etc.

### Common patterns in MJCF

- The robot is usually a **tree of bodies**: each `<body>` can contain child `<body>` nodes.
- Joints are defined *inside* the body they belong to; they determine how that body moves relative to its parent.
- Visual and collision shapes are `<geom>` elements attached to bodies.
- Actuators often reference joints (directly or via a transmission), enabling control inputs.

### Tips for reading an MJCF quickly

1. Start from `<worldbody>` to understand the scene and the body hierarchy.
2. Look for the robot base body, then follow nested `<body>` blocks to see limbs.
3. Scan `<joint>` tags to understand the DOFs (names and types).
4. Scan `<actuator>` to see what is controllable (and how many control inputs you have).
5. If something looks missing visually, check `<asset>` and `<geom type="mesh">` references.

## How to interact with the MuJoCo viewer

The simulation viewer supports **drag-and-drop interaction** with simulated objects to apply forces or torques.

- **Double-click** on a body to select it.
- **Ctrl + left drag** applies a **torque** to the selected object (rotation).
- **Ctrl + right drag** applies a **force** to the selected object in the **(x, z)** plane (translation).
- **Ctrl + Shift + right drag** applies a **force** to the selected object in the **(x, y)** plane.

## 1: Chest+Head yaw

```python
# ---- 1: Chest+Head yaw ----
xml_path = r"unitree_g1/g1_head_1dof.xml"
assert os.path.exists(xml_path), f"Missing file: /mnt/data/g1_head_1dof.xml"

with open(xml_path, "r", encoding="utf-8") as f:
    xml_text = f.read()

print("XML path:", xml_path)
print("Size (chars):", len(xml_text))

display(HTML(
    "<div style='border:1px solid #ddd; padding:10px; max-height:420px; overflow:auto; white-space:pre; font-family:monospace; font-size:12px;'>"
    + xml_text.replace("&","&amp;").replace("<","&lt;").replace(">","&gt;")
    + "</div>"
))

model = mujoco.MjModel.from_xml_path(xml_path)
data  = mujoco.MjData(model)

# Close the viewer window to proceed to the next cell.
mujoco.viewer.launch(model, data)
```
